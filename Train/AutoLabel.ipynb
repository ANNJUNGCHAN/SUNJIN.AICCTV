```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os
```

```python
MODEL_PATH = "/Drive/DATACENTER_SSD/AICCTV_ASSET/model/20240612_trial_1/weights/best.pt"
DATA_PATH = "/Drive/DATACENTER_HDD/AICCTV_RAW_VIDEO"
SAVE_PATH = "/Drive/DATACENTER_HDD/AICCTV_TRAIN_TEMP"
```

### 전체 추출

```python
# 비디오 리스트 추출하기
video_list = os.listdir(DATA_PATH)

# Load a modelrm 
model = YOLO(MODEL_PATH)

for video_path in video_list :
    
    if 'mp4' in video_path :
    
        video_name = video_path.replace(".mp4","")
    
    elif 'avi' in video_path :
        
        video_name = video_path.replace(".avi","")
    
    save_path = os.path.join(SAVE_PATH, video_name)
    
    os.makedirs(os.path.join(SAVE_PATH, video_name))
    
    # 비디오 하나 추출
    full_video_path = os.path.join(DATA_PATH, video_path)
    print("START : ", full_video_path)
    
    # 비디오 하나 돌리기
    results = model(
    source = full_video_path,
    save_txt = False,
    save = False,
    stream = True,
    verbose = False
    )
    
    frame = 0

    for result in results :
        
        save_name = video_name + "_" + str(frame).zfill(7)
        
        boxes = result.boxes.xywhn.cpu()
        clses = result.boxes.cls.cpu()
        img = result.orig_img
        
        if len(boxes) == 0:
            
            cv2.imwrite(os.path.join(save_path, save_name + ".jpg"), img)
            
            # 빈 텍스트 파일 생성하고 아무것도 하지 않음
            with open(os.path.join(save_path, save_name + ".txt"), 'w') as file:
                pass
            
            frame += 1
        
        else :
            
            cv2.imwrite(os.path.join(save_path, save_name + ".jpg"), img)
            
            with open(os.path.join(save_path, save_name + ".txt"), 'w') as file:
            
                for index in range(len(clses)) :
                    
                    class_id = int(clses[index].tolist())
                    boxes_list = boxes[index].tolist()
                    
                    line = f'{class_id} ' + ' '.join(f'{coord:.6f}' for coord in boxes_list) + '\n'
                    file.write(line)
                
            frame += 1
```
